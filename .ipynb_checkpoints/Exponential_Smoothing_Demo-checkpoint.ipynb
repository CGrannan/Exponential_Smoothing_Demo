```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
sns.set_context('talk')
import warnings
warnings.filterwarnings("ignore")

from statsmodels.tsa.seasonal import seasonal_decompose as sd
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
from statsmodels.tsa.holtwinters import Holt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
# Load data
df = pd.read_csv('monthly-beer-production-in-austr.csv')
df.Month = pd.to_datetime(df.Month)
data = df.set_index('Month')
data.head()
```

```python
fig = plt.figure(figsize=(14,6))
plt.plot(data)
plt.title('Monthly Beer Production')
plt.xlabel('Date')
plt.ylabel('Kegs Produced')
plt.show()
```

```python
decomposed = sd(data)
decomposed.plot()
plt.show()
```

```python
model = SimpleExpSmoothing(data)
fit = model.fit()
fcast = fit.forecast(60)

# plot
fig = plt.figure(figsize=(14,6))

plt.plot(data, color='black', label='Observed')

fcast.plot(color='red', label='Forecast')

plt.title('Simple Exponential Smoothing Predictions')
plt.xlabel('Date')
plt.ylabel('Kegs Produced')
plt.legend()
plt.show()
```

```python
model = Holt(data).fit()
model2 = Holt(data, damped=True).fit()

fcast = model.forecast(60)
fcast2 = model2.forecast(60)

print('Model 1 Parameters:', model.params)
print('Model 2 Parameters:', model2.params)

# plot
fig = plt.figure(figsize=(14,6))

plt.plot(data, color='black',)

fcast.plot(color='red', label='Linear Trend', legend=True)
fcast2.plot(color='green', label='Dampened Trend', legend=True)

plt.title('Double Exponential Smoothing Results')
plt.xlabel('Date')
plt.ylabel('Kegs Produced')
plt.show()
```

```python
model = ExponentialSmoothing(data, trend='add', seasonal='add', seasonal_periods=12).fit()

fcast = model.forecast(60)
fcast2 = model2.forecast(60)

# plot
fig = plt.figure(figsize=(14,8))
plt.plot(data, color='black')

fcast.plot(color='red', label='Additive Trend and Seasonality')

plt.title('Triple Exponential Smoothing Results')
plt.xlabel('Date')
plt.ylabel('Kegs Produced')
plt.legend()
plt.show()
```
